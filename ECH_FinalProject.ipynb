```python
#Solution Thermodynamics
#Rxn: HMD + SA −−→ Nylon − 6, 10 + H2O(l)
#Gibbs/Enthalpy Nylon: https://www.sciencedirect.com/science/article/pii/S2405844023082506#:~:text=The%20values%20of%20the%20thermodynamic,as%20fibers%20and%20engineering%20thermoplastics.
#Gibbs/Enthalpy HMD: https://www.chemeo.com/cid/45-944-4/1-6-Hexanediamine#ref-joback
#Gibbs/Enthalpy SA: https://www.chemeo.com/cid/37-335-9/Sebacic-acid-di-2-hexyl-ester
import numpy as np
import matplotlib.pyplot as plt
#Standard Gibbs Energies of Formation for Rxn [kJ/mol]
G_Nylon= -46.62
G_HMD= 132.54
G_SA= -338.36
G_H2O= -237.129
G_Rxn= round((G_Nylon+G_H2O) - (G_HMD + G_SA),2)
print(G_Rxn, "kJ/mol is the standard Gibbs reaction energy.")


R= 8.314   #J/ mol·K
T= 298.15 #K
#Value of Equilibrium Constant
#Equation 1, pp.2
K= round(np.exp(-G_Rxn/(R*T)),6)
print(K, "is the equilibrium constant for the formation of this reaction.")
#___________________________________________________________________________________________________________________________________________
#Want to optimize the temperature of the reaction to favor the product, and the formation of Nylon-6,10, rearrange equation [2] for K_prime

#Standard Enthalpy Value of Formation [kJ/mol]
H_Nylon= 40.25
H_HMD= -99.59
H_SA= -997.57
H_H20= -285.83
H_Rxn=(H_Nylon+H_H20) - (H_HMD + H_SA)
print(round(H_Rxn,2), "kJ/mol is the standard enthalpy of formation.")

#Graph of Temperature vs K'
T_prime= np.linspace(298.15, 550, 200)
K_prime= K* 1/(np.exp(-H_Rxn/(R*T)*(1/T-1/T_prime)))
#plotting
plt.plot(T_prime, K_prime)
plt.grid()
plt.xlabel('Temperature (K)')
plt.ylabel('Equilibrium Constant')
plt.title('Optimizing Equilibrium using Temperature')
plt.show()
```

```python
#Real Gibbs Energy of Mixing
import numpy as np
import matplotlib.pyplot as plt
#from a research paper, listed below
#https://citeseerx.ist.psu.edu/document?repid=rep1&type=pdf&doi=197b58d0549ff39e10ca095f8f6cfcb80f86a2ad

R= 8.314 # [=] J/ mol·K
#alpha and beta are calculated in the paper and help to calculate the Flory Huggins Interaction Parameter (FH)
alpha= 1.078
beta= -451.6
#volume fraction of Nylon
phi1= np.linspace(0,1,200)
#volume fraction of water, the solution
phi2= 1-phi1
T =np.linspace(300,550,200) #[=] K
FH= (1/R) * (alpha + beta/T)  #Flory Huggins Interaction Parameter
#Degrees of Polymerization of Nylon and Water respectively
N1=1
N2=1
#ideal
Gmix= (R*T)*( (phi1/N1)* np.log(phi1) + (phi2/N2)* np.log(phi2) + FH*phi1*phi2) 
#excess
G_excess = R*T*(FH * phi1 * phi2)
#real
G_real= (Gmix+ G_excess)/(R*T)

plt.plot(phi1, G_real)
plt.grid()
plt.xlabel('Volume Fraction of Nylon')
plt.ylabel('Real Gibbs Energy of Mixing/RT')
plt.title('Real Gibbs Energy of Mixing/RT vs Volume Fraction of Nylon')
plt.show()

```

```python
#Real Gibbs Energy of Mixing vs G ideal of mixing
import numpy as np
import matplotlib.pyplot as plt
#from a research paper, listed below
#https://citeseerx.ist.psu.edu/document?repid=rep1&type=pdf&doi=197b58d0549ff39e10ca095f8f6cfcb80f86a2ad

R= 8.314 # [=] J/ mol·K
#alpha and beta are calculated in the paper and help to calculate the Flory Huggins Interaction Parameter (FH)
alpha= 1.078
beta= -451.6
#volume fraction of Nylon
phi1= np.linspace(0,1,200)
#volume fraction of water, the solution
phi2= 1-phi1
T =np.linspace(300,550,200) #[=] K
FH= (1/R) * (alpha + beta/T)  #Flory Huggins Interaction Parameter
#Degrees of Polymerization of Nylon and Water respectively
N1=1
N2=1
#ideal
Gmix= (R*T)*( (phi1/N1)* np.log(phi1) + (phi2/N2)* np.log(phi2) + FH*phi1*phi2) 
Gmix_RT= Gmix/(R*T)
#excess
G_excess = R*T*(FH * phi1 * phi2)
#real
G_real= (Gmix+ G_excess)/(R*T)

#G real
plt.plot(phi1, G_real, label= 'G_Real/RT')
#G ideal
plt.plot(phi1, Gmix_RT, label = 'G_Ideal/RT')
plt.grid()
plt.legend()
plt.xlabel('Volume Fraction of Nylon')
plt.ylabel('Real Gibbs Energy of Mixing/RT')
plt.title('Real and Ideal Gibbs Energy of Mixing/RT vs Volume Fraction of Nylon')
plt.show()


#These print statements to visualize deviation in the real and ideal behavior
#Suchi high temperatures so behaves ideally.
print(G_real)
print(Gmix_RT)
```
